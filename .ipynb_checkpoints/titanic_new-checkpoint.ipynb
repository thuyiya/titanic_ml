## Data Cleaning and Handling Missing Values:

 - Handling Missing Ages
 - Handling Missing Embarked Values with most common value
 - Dropping Irrelevant Columns
 - Encoding Categorical Variables
 - Handling Missing Fare Values

```python
# Check for missing values in the dataset
print("Missing values before handling:")
print(df.isnull().sum())

# 1. Fill missing 'Age' values with median age per Passenger Class (Pclass)
# Reset index to avoid potential issues with groupby and alignment
df = df.reset_index(drop=True)

# Fill missing 'Age' values based on median of each 'Pclass'
df['Age'] = df.groupby('Pclass')['Age'].transform(lambda x: x.fillna(x.median()))

# 2. Fill missing 'Embarked' values with the most frequent embarkation point
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# 3. Drop 'Cabin' column due to high percentage of missing values
df.drop(columns=['Cabin'], inplace=True)

# 4. Convert categorical variables to numeric (Encoding 'Sex' and 'Embarked')
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# One-hot encode 'Embarked' column (dropping one to avoid multicollinearity)
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)

# 5. Handle missing 'Fare' values (if any), filling with median fare per class
df['Fare'].fillna(df.groupby('Pclass')['Fare'].transform('median'), inplace=True)

# Check for missing values after processing
print("Missing values after handling:")
print(df.isnull().sum())

# Display the cleaned dataset
print("Dataset after cleaning:")
print(df.head())
```

## Feature Engineering

```python
# 1. Extract titles from the Name column
df['Title'] = df['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())

# 2. Create binary features based on titles
df['Is_Mr'] = df['Title'].apply(lambda x: 1 if x == 'Mr' else 0)
df['Is_Mrs'] = df['Title'].apply(lambda x: 1 if x == 'Mrs' else 0)
df['Is_Miss'] = df['Title'].apply(lambda x: 1 if x == 'Miss' else 0)
df['Is_Master'] = df['Title'].apply(lambda x: 1 if x == 'Master' else 0)
df['Is_Dr'] = df['Title'].apply(lambda x: 1 if x == 'Dr' else 0)

# 3. Family size and IsAlone
df['Family_Size'] = df['SibSp'] + df['Parch']
df['IsAlone'] = df['Family_Size'].apply(lambda x: 1 if x == 0 else 0)

# 4. Fare per person
df['Fare_Per_Person'] = df['Fare'] / (df['Family_Size'] + 1)

# 5. Ticket length (length of the ticket string)
df['Ticket_Length'] = df['Ticket'].apply(lambda x: len(x))

# 6. Interaction between Age and Pclass
df['Age_x_Pclass'] = df['Age'] * df['Pclass']

# 7. Fare and Embarked interaction
df['Fare_x_Embarked'] = df['Fare'] * df['Embarked_S']
```
